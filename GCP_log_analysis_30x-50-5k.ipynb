```python
results = {}
for i in range(5001):
    results[i] = {}
print(results)
```

```python
trialcount = 0;
rowsList = []

for file in filenames:
    with open(os.path.join('./run_3_logs', file)) as json_file:
        for i, line in enumerate(json_file):
            x = json.loads(line)
            trialId = int(x['labels']['ml_googleapis_com_trial_id'])
#             if trialId == 4347:
#                 print(x)
            message = x['jsonPayload']['message']
            results[trialId]['trialId'] = trialId            
            if 'Fitness Score' in message:
                results[trialId]['fitnessScore'] = float(message.split()[3])
            if 'Fitness' in message:
                results[trialId]['fitness'] = float(message.split()[3])
            if 'RateOfLayingPheromone' in message:
                results[trialId]['RateOfLayingPheromone'] = float(message.split()[1])
            if 'ProbabilityOfSwitchingToSearching' in message:
                results[trialId]['ProbabilityOfSwitchingToSearching'] = float(message.split()[1])
            if 'ProbabilityOfReturningToNest' in message:
                results[trialId]['ProbabilityOfReturningToNest'] = float(message.split()[1])
            if 'RateOfInformedSearchDecay' in message:
                results[trialId]['RateOfInformedSearchDecay'] = float(message.split()[1])
            if ('RateOfSiteFidelity' in message):
                results[trialId]['RateOfSiteFidelity'] = float(message.split()[1])
            if ('RateOfPheromoneDecay' in message):
                results[trialId]['RateOfPheromoneDecay'] = float(message.split()[1])
            if ('UninformedSearchVariation' in message):
                results[trialId]['UninformedSearchVariation'] = float(message.split()[1])
#                 if ('food distribution' in message and 'randomizing' not in message):
#                     foodDistribution.append(message.split()[2])

for i in range(1,5001):
    rowsList.append(results[i])
# print(rowsList)
# print(results)
            
```

```python
df = pd.DataFrame(rowsList)
df.sort_values(by=['fitnessScore'], ascending=False)
```
