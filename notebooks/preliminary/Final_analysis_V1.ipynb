```python
import os, csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame(columns=['score','found', 'params', 'dist'])
df
```

```python
filenames = [f for f in os.listdir('./param_compares') if f[-3:] == 'txt']
```

```python
scores = []
founds = []
params = []
dists = []
    
for file in filenames:
    with open(os.path.join('./param_compares', file)) as textfile:
        param = file.split('_')[2][1:]
        dist = file.split('_')[3][1:2] 
        for line in textfile:
            found = int(float(line.strip().split()[0]))
            score = float(line.strip().split()[1])
            scores.append(score)
            founds.append(found)
            params.append(param)
            dists.append(dist)
df['score'] = scores
df['found'] = founds
df['params'] = params
df['dist'] = dists
df
            
#         for row in csvreader:

            
#         print(rowsList)
        
            
```

```python
df[(df['params'] == 'T') & (df['dist'] == '2')]['score'].describe()
```

```python
df = df.sort_values(by=['params', 'dist'])
df.to_csv('final_tally.csv')
df
```

```python
trial_p0d0 = df[(df['params'] == '0') & (df['dist'] == '0')]['score']
trial_p0d1 = df[(df['params'] == '0') & (df['dist'] == '1')]['score']
trial_p0d2 = df[(df['params'] == '0') & (df['dist'] == '2')]['score']

trial_p1d0 = df[(df['params'] == '1') & (df['dist'] == '0')]['score']
trial_p1d1 = df[(df['params'] == '1') & (df['dist'] == '1')]['score']
trial_p1d2 = df[(df['params'] == '1') & (df['dist'] == '2')]['score']

trial_p2d0 = df[(df['params'] == '2') & (df['dist'] == '0')]['score']
trial_p2d1 = df[(df['params'] == '2') & (df['dist'] == '1')]['score']
trial_p2d2 = df[(df['params'] == '2') & (df['dist'] == '2')]['score']

trial_pTd0 = df[(df['params'] == 'T') & (df['dist'] == '0')]['score']
trial_pTd1 = df[(df['params'] == 'T') & (df['dist'] == '1')]['score']
trial_pTd2 = df[(df['params'] == 'T') & (df['dist'] == '2')]['score']
```

```python
fig1, ax1 = plt.subplots(figsize=(5, 2))
ax1.set_title('Uniform Food Distribution (0)')
plt.xlabel('behavioral param set', fontweight='bold', fontsize=12, labelpad=20,)
plt.ylabel('score', fontweight='bold', fontsize=12)
ax1.boxplot([trial_p0d0,trial_p1d0,trial_p2d0,trial_pTd0])
ax1.set_xticklabels(np.repeat(["GA Uniform","GA Cluster","GA Power Law", "HyperTune"], 1),rotation=45, fontsize=10, ha="right")
```
